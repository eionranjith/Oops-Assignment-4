```python
#Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.
class BankAccount:
    def __init__(self,account_number,account_holder_name,balance=0):
        self.account_number=account_number
        self.account_holder_name=account_holder_name
        self.balance=balance
    def deposit(self,amount):
        if amount>0:
            self.balance +=amount
            print(f"Deposited {amount} dollars into account {self.account_number}.")
        else:
            print("Invalid amount for deposit.")
    def withdraw(self,amount):
        if 0<amount<=self.balance:
            self.balance -=amount
            print(f"withdrew {amount} dollars from account {self.account_number}.")
        else:
            print("Insufficient funds or invalid amount for withdrawal.")
    def display_balance(self):
        print(f"account {self.account_number} balance : {self.balance} dollars.")
account1=BankAccount("123456789","Ranjith",1000)
account1.display_balance()
account1.deposit(500)
account1.display_balance()
account1.withdraw(200)
account1.display_balance()
account1.withdraw(2000)
```

```python
#3)Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}

    def rent_vehicle(self, vehicle_type, vehicle_id):
        if vehicle_type in self.available_vehicles:
            if vehicle_id in self.available_vehicles[vehicle_type]:
                del self.available_vehicles[vehicle_type][vehicle_id]
                print(f"Vehicle {vehicle_id} of type {vehicle_type} rented successfully.")
            else:
                print(f"Vehicle {vehicle_id} of type {vehicle_type} is not available.")
        else:
            print(f"No vehicles of type {vehicle_type} available for rent.")

    def return_vehicle(self, vehicle_type, vehicle_id):
        if vehicle_type in self.available_vehicles:
            if vehicle_id not in self.available_vehicles[vehicle_type]:
                self.available_vehicles[vehicle_type][vehicle_id] = True
                print(f"Vehicle {vehicle_id} of type {vehicle_type} returned successfully.")
            else:
                print(f"Vehicle {vehicle_id} of type {vehicle_type} is already available.")
        else:
            self.available_vehicles[vehicle_type] = {vehicle_id: True}
            print(f"Vehicle {vehicle_id} of type {vehicle_type} returned successfully.")

    def display_available_vehicles(self):
        if not self.available_vehicles:
            print("No vehicles available.")
        else:
            print("Available Vehicles:")
            for vehicle_type, vehicles in self.available_vehicles.items():
                print(f"{vehicle_type}: {', '.join(vehicles.keys())}")


rental_system = VehicleRentalSystem()
rental_system.display_available_vehicles()

rental_system.return_vehicle("Car", "C001")
rental_system.display_available_vehicles()

rental_system.available_vehicles = {"Car": {"C001": True, "C002": True}, "Bike": {"B001": True}}
rental_system.display_available_vehicles()

rental_system.rent_vehicle("Car", "C001")
rental_system.display_available_vehicles()

rental_system.rent_vehicle("Car", "C003")
```

```python
#Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.?
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.borrowed = False

    def borrow(self):
        if not self.borrowed:
            self.borrowed = True
            return True
        else:
            print("Sorry, this book is already borrowed.")
            return False

    def return_book(self):
        if self.borrowed:
            self.borrowed = False
            return True
        else:
            print("This book is not borrowed.")
            return False

    def __str__(self):
        return f"Title: {self.title}, Author: {self.author}, Borrowed: {self.borrowed}"


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title:
                return book.borrow()
        print("Book not found.")
        return False

    def return_book(self, title):
        for book in self.books:
            if book.title == title:
                return book.return_book()
        print("Book not found.")
        return False

    def display_available_books(self):
        available_books = [book for book in self.books if not book.borrowed]
        if available_books:
            print("Available Books:")
            for book in available_books:
                print(book)
        else:
            print("No books available.")



library = Library()

book1 = Book("Harry Potter and the Philosopher's Stone", "J.K. Rowling")
book2 = Book("1984", "George Orwell")
book3 = Book("To Kill a Mockingbird", "Harper Lee")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.display_available_books()

library.borrow_book("Harry Potter and the Philosopher's Stone")
library.borrow_book("1984")

library.display_available_books()

library.return_book("1984")

library.display_available_books()
```

```python
#Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.?
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        self.quantity = new_quantity

    def __str__(self):
        return f"Name: {self.name}, Price: ${self.price}, Quantity: {self.quantity}"


class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def update_product_quantity(self, product_name, new_quantity):
        for product in self.products:
            if product.name == product_name:
                product.update_quantity(new_quantity)
                print(f"Quantity of {product_name} updated to {new_quantity}.")
                return
        print("Product not found.")

    def display_available_products(self):
        if self.products:
            print("Available Products:")
            for product in self.products:
                print(product)
        else:
            print("No products available.")



inventory = Inventory()

product1 = Product("Laptop", 1000, 5)
product2 = Product("Smartphone", 800, 10)
product3 = Product("Headphones", 100, 20)

inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)

inventory.display_available_products()

inventory.update_product_quantity("Laptop", 3)
inventory.display_available_products()

inventory.update_product_quantity("Smartphone", 15)
inventory.display_available_products()

inventory.update_product_quantity("Tablet", 8)
```

```python
#Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.?
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def area(self):
        raise NotImplementedError("Area calculation not implemented for the base Shape class")

    def perimeter(self):
        raise NotImplementedError("Perimeter calculation not implemented for the base Shape class")

class Rectangle(Shape):
    def __init__(self, length, width):
        super().__init__(length, width)

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)

class Triangle(Shape):
    def __init__(self, side1, side2, side3):
        super().__init__(side1, side2, side3)

    def area(self):
        # Using Heron's formula for the area of a triangle
        s = (self.length + self.width + self.height) / 2
        return (s * (s - self.length) * (s - self.width) * (s - self.height)) ** 0.5

    def perimeter(self):
        return self.length + self.width + self.height


rectangle = Rectangle(5, 4)
print("Rectangle Area:", rectangle.area())
print("Rectangle Perimeter:", rectangle.perimeter())

square = Square(5)
print("Square Area:", square.area())
print("Square Perimeter:", square.perimeter())

triangle = Triangle(3, 4, 5)
print("Triangle Area:", triangle.area())
print("Triangle Perimeter:", triangle.perimeter())
```

```python
#Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.?
class Student:
    def __init__(self,student_id,name):
        self.student_id=student_id
        self.name=name
        self.grades=[]
    def add_grade(self,grade):
        self.grades.append(grade)
    def calculate_average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades)/len(self.grades)
    def display_details(self):
        print("Student ID:",self.student_id)
        print("Name:",self.name)
        print("Grades:",self.grades)
        print("Average Grade:",self.calculate_average_grade())
student1=Student("13110350","Ranjith")
student1.add_grade(85)
student1.add_grade(90)
student1.add_grade(75)
student1.display_details()
```

```python
#Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.?
class SocialMediaProfile:
    def __init__(self,username):
        self.username=username
        self.posts=[]
    def add_post(self,post):
        self.posts.append(post)
    def display_posts(self):
        if not self.posts:
            print("No posts available.")
        else:
            print(f"Posts by {self.username}:")
            for post in self.posts:
                print(post)
    def search_posts(self,keyword):
        found_posts=[]
        for post in self.posts:
            if keyword.lower() in post.lower():
                found_posts.append(post)
            if found_posts:
                print(f"posts containing '{keyword}' by {self.username}:")
                for post in found_posts:
                    print(post)
            else:
                print(f"No posts containing '{keyword}' found by {self.username}.")
profile1=SocialMediaProfile("user1")
profile1.add_post("Hello world!")
profile1.add_post("This is my first post.")
profile1.add_post("Just enjoying the day.")
profile1.display_posts()
profile1.search_posts("enjoying")


profile2=SocialMediaProfile("user2")
profile2.add_post("Today is a great day!")
profile2.add_post("Feeling excited about my new project.")
profile2.add_post("lunchtime!")
profile2.display_posts()
profile2.search_posts("excited")
        
```

```python
#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.?
class ToDoList:
    def __init__(self):
        self.tasks={}
    def add_task(self,task,due_date):
        self.tasks[task]=due_date
    def mark_as_completed(self,task):
        if task in self.tasks:
            del self.tasks[task]
            print(f"Task '{task}' marked as completed.")
        else:
            print(f"Task '{task}' not found.")
    def display_pending_tasks(self):
        if not self.tasks:
            print("no Pending tasks.")
        else:
            print("Pending tasks.")
            for task,due_date in self.tasks.items():
                print(f"{task} - due on :{due_date}")
todo_list=ToDoList()
todo_list.add_task("Complete project proposal","2024-03-10")
todo_list.add_task("Buy groceries ","2024-03-06")
todo_list.add_task("Call mom","2024-03-08")

todo_list.display_pending_tasks()

todo_list.mark_as_completed("Buy groceries")
todo_list.display_pending_tasks()

todo_list.mark_as_completed("write report")                      
                      
            
```
